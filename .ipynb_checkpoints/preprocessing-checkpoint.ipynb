```python
# import libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# load data 
data = pd.read_csv('./diabetic_data.csv')
```

```python
# split data into X and y 
X = data.drop('readmitted', axis = 1)
y = data['readmitted']
```

```python
# get counts of missing values 
missing_values = (X == '?').sum(axis = 0)

# only keep features that have at least 1 missing value 
missing_values_filtered = missing_values[missing_values > 0]

# plot missing value counts 
plt.figure(figsize = (15, 8))

plt.bar(missing_values_filtered.index, missing_values_filtered.values / len(data))
plt.xlabel('Feature', size = 20)
plt.ylabel('Fraction of Missing Values', size = 20)
plt.title('Fraction of Missing Values in the Features', size = 20)

plt.xticks(rotation = 30, size = 20)
plt.yticks(size = 20)

plt.show()
```

```python
# drop columns that we don't need 
columns_to_drop = ['encounter_id', 'patient_nbr', 'weight', 'payer_code', 'medical_specialty', 'examide', 'citoglipton']

X = X.drop(columns_to_drop, axis = 1)
```

```python
# split the data into dev and test set 
from sklearn.model_selection import train_test_split

X_dev, X_test, y_dev, y_test = train_test_split(X, y, stratify = y, test_size = 0.2)
X_train, X_val, y_train, y_val = train_test_split(X_dev, y_dev, stratify = y_dev, test_size = 0.2)
```

```python
# impute missing values for categorical variables 
from sklearn.impute import SimpleImputer 

feature_names = X.columns

imp = SimpleImputer(missing_values = '?', strategy = 'most_frequent')
X_train = imp.fit_transform(X_train)
X_train = pd.DataFrame(X_train, columns = feature_names)

X_val = imp.transform(X_val)
X_val = pd.DataFrame(X_val, columns = feature_names)

X_test = imp.transform(X_test)
X_test = pd.DataFrame(X_test, columns = feature_names)
```

```python
# label encode target variable
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

y_train = le.fit_transform(y_train)
y_train = pd.Series(y_train)

y_val = le.transform(y_val)
y_val = pd.Series

y_test = le.transform(y_test)
y_test = pd.Series(y_test)
```

```python
# one hot encode target variable for target encoding of other variables 
from sklearn.preprocessing import OneHotEncoder 

ohe = OneHotEncoder(drop = 'first')

y_train_ohe = ohe.fit_transform(y_train.to_numpy().reshape(-1, 1)).toarray()
y_test_ohe = ohe.transform(y_test.to_numpy().reshape(-1, 1)).toarray()

col_names = ['target_1', 'target_2']

y_train_ohe = pd.DataFrame(y_train_ohe, columns = col_names)
y_test_ohe = pd.DataFrame(y_test_ohe, columns = col_names)
```

```python
# target encode categorical features that aren't ordinal 
from category_encoders import TargetEncoder

te_features = ['race', 'gender', 'admission_type_id', 'discharge_disposition_id', 'admission_source_id',
              'diag_1', 'diag_2', 'diag_3', 'max_glu_serum', 'A1Cresult', 'metformin', 
              'repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride', 'acetohexamide', 
              'glipizide', 'glyburide', 'tolbutamide', 'pioglitazone', 'rosiglitazone', 
              'acarbose', 'miglitol', 'troglitazone', 'tolazamide', 
              'insulin', 'glyburide-metformin', 'glipizide-metformin', 'glimepiride-pioglitazone', 
              'metformin-rosiglitazone', 'metformin-pioglitazone', 'change', 'diabetesMed']

X_train_te = X_train[te_features]
X_test_te = X_test[te_features]

X_train_encoded = X_train
X_test_encoded = X_test

for c in range(len(col_names)): 
    te = TargetEncoder()
    
    temp = te.fit_transform(X_train_te, y_train_ohe[col_names[c]])
    temp.columns = [s + '_' + str(c + 1) for s in temp.columns]
    X_train_encoded = pd.concat([X_train_encoded, temp], axis = 1)
    
    temp = te.transform(X_test_te, y_test_ohe[col_names[c]])
    temp.columns = [s + '_' + str(c + 1) for s in temp.columns]
    X_test_encoded = pd.concat([X_test_encoded, temp], axis = 1)
    
X_train_encoded = X_train_encoded.drop(te_features, axis = 1)
X_test_encoded = X_test_encoded.drop(te_features, axis = 1)
```

```python
# ordinal encode the age column 
from sklearn.preprocessing import OrdinalEncoder 

oe = OrdinalEncoder()

X_train_encoded['age'] = oe.fit_transform(X_train_encoded['age'].to_numpy().reshape(-1, 1))
X_test_encoded['age'] = oe.transform(X_test_encoded['age'].to_numpy().reshape(-1, 1))
```

```python
# scale the data using standard scaler 
from sklearn.preprocessing import StandardScaler 

ss = StandardScaler()

X_train_encoded = ss.fit_transform(X_train_encoded)
X_test_encoded = ss.transform(X_test_encoded)
```

```python
pd.DataFrame(X_train_encoded).to_csv('X_train_encoded.csv')
pd.DataFrame(X_test_encoded).to_csv('X_test_encoded.csv')
pd.Series(y_train).to_csv('y_train_encoded.csv')
pd.Series(y_test).to_csv('y_test_encoded.csv')
```
